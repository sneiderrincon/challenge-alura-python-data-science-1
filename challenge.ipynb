```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



## **INGRESO TOTAL CADA TIENDA**



### En este primer análisis, debes calcular el ingreso total de cada tienda. Sumando los valores de la columna Precio de cada conjunto de datos de la tienda para estimar los ingresos.

```python
ingreso_total_tienda1 = tienda['Precio'].sum()
ingreso_total_tienda2 = tienda2['Precio'].sum()
ingreso_total_tienda3 = tienda3['Precio'].sum()
ingreso_total_tienda4 = tienda4['Precio'].sum()

print(f"Ingreso total Tienda 1: {ingreso_total_tienda1}")
print(f"Ingreso total Tienda 2: {ingreso_total_tienda2}")
print(f"Ingreso total Tienda 3: {ingreso_total_tienda3}")
print(f"Ingreso total Tienda 4: {ingreso_total_tienda4}")
```

```python
import matplotlib.pyplot as plt
import pandas as pd

ingresos_totales = pd.DataFrame({
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Ingreso Total': [ingreso_total_tienda1, ingreso_total_tienda2, ingreso_total_tienda3, ingreso_total_tienda4]
})

plt.figure(figsize=(8, 5))
plt.plot(ingresos_totales['Tienda'], ingresos_totales['Ingreso Total'], marker='o', linestyle='-', color='purple')
plt.xlabel('Tienda')
plt.ylabel('Ingreso Total')
plt.title('Ingreso Total por Tienda')
plt.ylim(ingresos_totales['Ingreso Total'].min() * 0.9, ingresos_totales['Ingreso Total'].max() * 1.1) # Adjust y-axis limits
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Análisis de Ingreso Total por Tienda

Observando el gráfico de línea que muestra el ingreso total por tienda, podemos identificar las siguientes tendencias:

*   La **Tienda 1** tiene el mayor ingreso total, seguida por la **Tienda 2**, **Tienda 3**, y finalmente la **Tienda 4** con el menor ingreso total.
*   Hay una disminución gradual en el ingreso total de la Tienda 1 a la Tienda 4.

Este gráfico permite comparar visualmente el rendimiento financiero de cada tienda en términos de ingreso total.

```python
import matplotlib.pyplot as plt

# Function to create a pie chart for a given store
def plot_pie_chart(data, title):
    plt.figure(figsize=(8, 8))
    plt.pie(data, labels=data.index, autopct='%1.1f%%', startangle=140)
    plt.title(title)
    plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
    plt.show()

# Plot pie charts for each store
plot_pie_chart(ventas_por_categoria_tienda1, 'Ventas por Categoría Tienda 1')
plot_pie_chart(ventas_por_categoria_tienda2, 'Ventas por Categoría Tienda 2')
plot_pie_chart(ventas_por_categoria_tienda3, 'Ventas por Categoría Tienda 3')
plot_pie_chart(ventas_por_categoria_tienda4, 'Ventas por Categoría Tienda 4')
```

# 3. Calificación promedio de la tienda


### En este paso, debes calcular las calificaciones promedio de los clientes para cada tienda. El objetivo es conocer la satisfacción del cliente con los productos vendidos.

```python
calificacion_promedio_tienda1 = tienda['Calificación'].mean()
calificacion_promedio_tienda2 = tienda2['Calificación'].mean()
calificacion_promedio_tienda3 = tienda3['Calificación'].mean()
calificacion_promedio_tienda4 = tienda4['Calificación'].mean()
```

```python
print(f"Calificación promedio Tienda 1: {calificacion_promedio_tienda1}")
print(f"Calificación promedio Tienda 2: {calificacion_promedio_tienda2}")
print(f"Calificación promedio Tienda 3: {calificacion_promedio_tienda3}")
print(f"Calificación promedio Tienda 4: {calificacion_promedio_tienda4}")
```

```python
import matplotlib.pyplot as plt
import pandas as pd

calificaciones_promedio = pd.DataFrame({
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Calificación Promedio': [calificacion_promedio_tienda1, calificacion_promedio_tienda2, calificacion_promedio_tienda3, calificacion_promedio_tienda4]
})

plt.figure(figsize=(8, 5))
plt.bar(calificaciones_promedio['Tienda'], calificaciones_promedio['Calificación Promedio'], color='teal')
plt.xlabel('Tienda')
plt.ylabel('Calificación Promedio')
plt.title('Calificación Promedio por Tienda')
plt.ylim(0, 5) # Ratings are typically on a scale of 1-5
plt.tight_layout()
plt.show()
```

## Análisis de Calificación Promedio por Tienda

Observando el gráfico de barras que muestra la calificación promedio por tienda, podemos notar que las calificaciones promedio de todas las tiendas son bastante similares, oscilando entre aproximadamente 3.97 y 4.05.

*   La **Tienda 3** tiene la calificación promedio ligeramente más alta (alrededor de 4.05).
*   La **Tienda 1** tiene la calificación promedio ligeramente más baja (alrededor de 3.97).

Aunque hay pequeñas variaciones, la cercanía de las calificaciones promedio sugiere que la satisfacción general del cliente es relativamente consistente en las cuatro tiendas, basándonos en esta métrica. Combine the latitude and longitude data from all stores into a single DataFrame and ensure the data types are appropriate for plotting.



```python
# Create a list of DataFrames
list_of_dfs = [tienda, tienda2, tienda3, tienda4]

# Concatenate the DataFrames
combined_df = pd.concat(list_of_dfs)

# Select 'lat' and 'lon' columns
df_geo = combined_df[['lat', 'lon']]

# Convert columns to numeric, coercing errors
df_geo['lat'] = pd.to_numeric(df_geo['lat'], errors='coerce')
df_geo['lon'] = pd.to_numeric(df_geo['lon'], errors='coerce')

# Display the first few rows and info of the new DataFrame
display(df_geo.head())
display(df_geo.info())
```

## Identificar puntos críticos de ventas

### Subtarea:

Analizar las visualizaciones para identificar las áreas con una alta concentración de ventas.


```python
import folium

# Calculate the mean of the latitude and longitude
mean_lat = df_geo['lat'].mean()
mean_lon = df_geo['lon'].mean()

# Create a Folium map centered around the mean coordinates
sales_map = folium.Map(location=[mean_lat, mean_lon], zoom_start=5)

# Add a Tile layer
folium.TileLayer('OpenStreetMap').add_to(sales_map)

# Add markers for each sale location
for index, row in df_geo.iterrows():
    folium.Marker([row['lat'], row['lon']]).add_to(sales_map)

# Display the map
display(sales_map)
```
